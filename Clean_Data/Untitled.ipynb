```python
nhl_08.head()
winners = []
losers = []
nhl_08
```

```python
score_diff = nhl_08["score_diff"].to_list()
home_team = nhl_08['Home'].to_list()
visitors = nhl_08['Visitor'].to_list()
len(score_diff)
```

```python
winners = []
losers = []
location = []
nhl_zip = zip(home_team, score_diff, visitors)
for home, score, visitor in nhl_zip:
    if score > 0:
        winners.append(home)
        losers.append(visitor)
        location.append(home)
    elif score < 0:
        losers.append(home)
        winners.append(visitor)
        location.append(home)
    game_dict = {"Winners":winners,
                "Losers": losers,
                "Location": location
                }
```

```python
len(winners)
```

```python
df_08 = pd.DataFrame(game_dict)
df_08
```
