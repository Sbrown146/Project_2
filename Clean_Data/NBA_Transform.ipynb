```python
def create_wl_df(df, records):
    df_count = records.groupby(["Winners"]).count()
    wl_series = (df_count["Losers"]/82*100).round(2)
    df_count["W/L Percentage"] = wl_series
    
    score_diff = df["score_diff"].to_list()
    home_team = df['Home/Neutral'].to_list()
    visitors = df['Visitor/Neutral'].to_list()
    
    winner = []
    nba_zip = zip(home_team, score_diff, visitors)
    for home, score, visitor in nba_zip:
        if score > 0:
            winner.append(home)
    
    
    home_wins = ((pd.Series(winner).value_counts())/41*100).round(2)

    df_count["Home W/L"] = home_wins
    
    df_count = df_count.drop(["Losers","Location"], axis = 1)
    
    return df_count
```
