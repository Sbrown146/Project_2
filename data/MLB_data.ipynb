```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
#path to csv
file= "MLB_results.csv"

#read data into pandas
MLB_df = pd.read_csv(file)
MLB_df.head(10)
```

```python
MLB_df.columns
```

```python
#EXTRACT "Team, Opposition, W/L, Runs and O-Runs"
reduced_MLB = MLB_df.loc[:, [ "Tm", "R","H/A","Opp","RA", "W/L","year"]]
reduced_MLB.head(15)
```

```python
reduced_MLB["H/A"]= reduced_MLB["H/A"].str.replace("@",'1', regex=True)
reduced_MLB.head(15)
```

```python
reduced_MLB= reduced_MLB.fillna(0)
reduced_MLB["H/A"]= reduced_MLB["H/A"].astype(int)
reduced_MLB.head(10)
```

```python
reduced_MLB['score_diff'] = reduced_MLB['R'] - reduced_MLB['RA']
reduced_MLB.head()
```

```python
reduced_MLB['Tm']
reduced_MLB.Tm.unique()
```

```python
reduced_MLB.rename(columns={"H/A": "HomeOrAway"}, inplace =True)
reduced_MLB.head()
```

```python
home_wins_df= pd.DataFrame()
for year in reduced_MLB['year'].unique().tolist():
    home_df2 = reduced_MLB.loc[(reduced_MLB['year']==year)\
                              &(reduced_MLB['HomeOrAway']==0),:].copy()
    
    home_wins2 = home_df2.loc[home_df2['score_diff']>0,'Tm'].values.tolist()
    home_wins_df[year] = pd.Series(dict(Counter(home_wins2)))
home_wins_df.fillna(0,inplace=True)
home_wins_df
```

```python
away_wins_df= pd.DataFrame()
for year in reduced_MLB['year'].unique().tolist():
    away_df2 = reduced_MLB.loc[(reduced_MLB['year']==year)\
                              &(reduced_MLB['HomeOrAway']==1),:].copy()
    
    away_wins2 = away_df2.loc[away_df2['score_diff']>0,'Tm'].values.tolist()
    away_wins_df[year] = pd.Series(dict(Counter(away_wins2)))
away_wins_df.fillna(0,inplace=True)
away_wins_df
```
