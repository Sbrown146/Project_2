```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
#path to csv
file= "NFL_results.csv"

#read data into pandas
NFL_df = pd.read_csv(file)
NFL_df.head(10)
```

```python
NFL_df.columns
```

```python
#EXTRACT "Team, Opposition, W/L, Runs and O-Runs"
reduced_NFL = NFL_df.loc[:, [ "season","Week", "Winner/tie","Var.7","Loser/tie","PtsW", "PtsL"]]
reduced_NFL.head(15)
```

```python
reduced_NFL.dtypes
```

```python
reduced_NFL['Week'] = reduced_NFL.Week.astype(str)
```

```python
reduced_NFL.dtypes
```

```python
reduced_nfl = reduced_NFL[~reduced_NFL['Week'].isin(["WildCard","Division","ConfChamp","SuperBowl","nan"])] 
```

```python
# reduced_NFL["Var.7"]= reduced_NFL["Var.7"].str.replace("@",'1', regex=True)
reduced_nfl
```

```python
# 0 means home, 1 means away
reduced_nfl= reduced_nfl.fillna(0)
# reduced_NFL["Var.7"]= reduced_NFL["Var.7"].astype(int)
reduced_nfl.head(10)
```

```python
reduced_nfl['score_diff'] = reduced_nfl['PtsW'] - reduced_nfl['PtsL']
reduced_nfl.head()
```

```python
reduced_nfl.rename(columns={'Winner/tie': "Winner", 'Var.7': "HomeOrAway"}, inplace =True)
reduced_nfl.head()
```

```python
home_wins_df= pd.DataFrame()
for season in reduced_nfl['season'].unique().tolist():
    home_df2 = reduced_nfl.loc[(reduced_nfl['season']==season)\
                              &(reduced_nfl['HomeOrAway']==0),:].copy()
    
    home_wins2 = home_df2.loc[:,'Winner'].values.tolist()
    home_wins_df[season] = pd.Series(dict(Counter(home_wins2)))
home_wins_df.fillna(0,inplace=True)
home_wins_df
```

```python
home_wins_df.div(8)
```

```python

away_wins_df= pd.DataFrame()
for season in reduced_nfl['season'].unique().tolist():
    away_df2 = reduced_nfl.loc[(reduced_nfl['season']==season)\
                              &(reduced_nfl['HomeOrAway']=='1'),:].copy()
#     print(away_df2.head())
    away_wins2 = away_df2.loc[:,'Winner'].values.tolist()
    away_wins_df[season] = pd.Series(dict(Counter(away_wins2)))
away_wins_df.fillna(0,inplace=True)
away_wins_df
```

```python
away_wins_df.div(8)
```
